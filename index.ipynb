## Step 1: Read the dataset and inspect its columns and 5-point statistics

```python
# Load necessary libraries and import the data
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn')

df = pd.read_csv('advertising.csv')
```

```python
# Check the columns and first few rows
df = df.drop('Unnamed: 0', axis=1)
df.head()
```

```python
# Get the 5-point statistics for data 
df.describe()
```

```python
# Describe the contents of this dataset
# The dataset consists of 200 rows and four columns (all floats)
```

## Step 2: Plot histograms with kde overlay to check the distribution of the predictors

```python
# For all the variables, check distribution by creating a histogram with kde
for column in df:
    df[column].plot.hist(density=True, label = column+' histogram')
    df[column].plot.kde(label =column+' kde')
    plt.legend()
    plt.show()
```

```python
# Record your observations here 
# Sales appears nearly normally distributed with slight left skew
# Radio and newspaper appear to be skewed to the left
# TV appears to be skewed to the right
```


## Step 3: Test for the linearity assumption 

Use scatterplots to plot each predictor against the target variable

```python
# visualize the relationship between the preditors and the target using scatterplots
fig, axs = plt.subplots(1, 3, sharey=True, figsize=(18, 6))
for idx, channel in enumerate(['TV', 'radio', 'newspaper']):
    df.plot(kind='scatter', x=channel, y='sales', ax=axs[idx], label=channel)
plt.legend()
plt.show()
```

```python
# Record yor observations on linearity here
# There appears to be a strong positive linear correlation between TV and sales
# There appears to be a weak positive linear correlation between radio and sales
# There does not appear to be a linear relationship between newspaper and sales
```

### Conclusion so far

Based on above initial checks, we can confidently say that TV and radio appear to be good predictors for our regression analysis. Newspaper is very heavily skewed and also doesnt show any clear linear relationship with the target. 
> We'll move ahead with our analysis using TV and radio, and rule out newspaper because we believe it violates OLS assumptions

Note: Kurtosis can be dealt with using techniques like log normalization to "push" the peak towards the center of distribution. You'll learn about this later on.

## Step 4: Run a simple regression in Statsmodels with TV as a predictor

```python
# import libraries
import statsmodels.api as sm
import statsmodels.formula.api as smf

# build the formula 
f = 'sales~TV'

# create a fitted model in one line
model = smf.ols(formula=f, data=df).fit()
```

## Step 5: Get Regression Diagnostics Summary

```python
model.summary()
# The R-squared value of 0.612 tells us that 61.2% of the variance
# in sales can be predicted by TV
```

Note here that the coefficients represent associations, not causations

## Step 6:  Draw a prediction line with data points on a scatter plot for X (TV) and Y (Sales)

Hint: You can use the `model.predict()` function to predict the start and end point of of regression line for the minimum and maximum values in the 'TV' variable. 

```python
# create a DataFrame with the minimum and maximum values of TV
tv_df = pd.DataFrame({"TV":[df.TV.min(), df.TV.max()]})
tv_df.head()

# make predictions for those x values and store them
pred = model.predict(tv_df)
pred

# first, plot the observed data and the least squares line
plt.plot(tv_df, pred, c="lime")
plt.scatter(df.TV,df.sales)
```

## Step 7: Visualize the error term for variance and heteroscedasticity

```python
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model, "TV", fig=fig)
plt.show()
```

```python
# Record Your observations on heteroscedasticity
# The Y and Fitted vs. X graph plots sales against our predicted 
# values with a confidence interval. The positive relationship shows 
# that TV and sales are correlated.

# The Residuals versus TV graph shows our model's errors versus the specified 
# predictor variable. Each dot is an observed value; the line represents the mean 
# of those observed values. The distance between the dots and the mean value 
# increases with TV expenditures, suggesting that the OLS assumption of homoskedasticity 
# does not hold.

# The Partial regression plot shows the relationship between TV spots and sales, 
# taking in to account the impact of adding other independent variables on our 
# existing TV coefficient.

# The Component and Component Plus Residual (CCPR) plot is an extension of the partial 
# regression plot. It shows where the trend line would lie after adding the impact of 
# other independent variables on the sales.
```

## Step 8: Check the normality assumptions by creating a QQ-plot

```python
# Code for QQ-plot here
import scipy.stats as stats
residuals = model.resid
fig = sm.graphics.qqplot(residuals, dist=stats.norm, line='45', fit=True, c='lightblue')
# fig.show()
```

```python
# Record Your observations on the normality assumption
# While the residuals do not seem to match up perfectly 
# with the plotted line, there seem to a number of
# deviations at the extremes. We cannot assume normality.
```

## Step 9: Repeat the above for radio and record your observations

```python
# build the formula 
f = 'sales~radio'

# create a fitted model in one line
model = smf.ols(formula=f, data=df).fit()
```

```python
# Record Your observations on heteroscedasticity
# The Y and Fitted vs. X graph plots sales against our predicted 
# values with a confidence interval. The positive relationship shows 
# that radio and sales are more or less correlated.

# The Residuals versus radio graph shows our model's errors versus the specified 
# predictor variable. Each dot is an observed value; the line represents the mean 
# of those observed values. The distance between the dots and the mean value 
# increases with radio expenditures, suggesting that the OLS assumption of homoskedasticity 
# does not hold.

# The Partial regression plot shows the relationship between radio spots and sales, 
# taking in to account the impact of adding other independent variables on our 
# existing radio coefficient.

# The Component and Component Plus Residual (CCPR) plot is an extension of the partial 
# regression plot. It shows where the trend line would lie after adding the impact of 
# other independent variables on sales.
```

```python
# Code for QQ-plot here
import scipy.stats as stats
residuals = model.resid
fig = sm.graphics.qqplot(residuals, dist=stats.norm, line='45', fit=True, c='lightblue')
# fig.show()
# Record your observations on the normality assumption
# The residuals do not seem to match up with the plotted line.
# We cannot assume normality.
```
